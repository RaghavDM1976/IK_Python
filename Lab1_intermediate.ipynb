```python
from strands.models.gemini import GeminiModel
from strands import Agent, tool
import os
from google.colab import userdata # Import userdata to retrieve the API key

@tool
def file_read(file_path: str) -> str:
    """Read a file and return its content."""
    # Enter your code here
    # Example implementation:
    try:
        with open(file_path, 'r') as f:
            return f.read()
    except FileNotFoundError:
        return f"Error: File not found at {file_path}"
    except Exception as e:
        return f"Error reading file {file_path}: {e}"

@tool
def file_write(file_path: str, content: str) -> str:
    """Write content to a file."""
    # Enter your code here
    # Example implementation:
    # with open(file_path, 'w') as f:
    #     f.write(content)
    # return f"Content written to {file_path}"
    return "File writing is not implemented in this example."

@tool
def list_directory(directory_path: str = ".") -> str:
    """List files and directories in the specified path."""
    # Enter your code here
    # Example implementation:
    try:
        return '\n'.join(os.listdir(directory_path))
    except FileNotFoundError:
        return f"Error: Directory not found at {directory_path}"
    except Exception as e:
        return f"Error listing directory {directory_path}: {e}"

# Retrieve the API key from Colab secrets
api_key = userdata.get('GOOGLE_API_KEY')

# Set the API key as an environment variable for the strands library to pick up automatically
os.environ['GOOGLE_API_KEY'] = api_key

# Configure the Gemini model
gemini_model = GeminiModel(
    model_id="gemini-flash-latest", # Changed from "gemini-1.5-flash-latest"
    params={
        "temperature": 0.7,
        "top_p": 0.9,
    },
)

system_prompt="You are a helpful assistant that provides concise responses."

# Create the agent with tools
local_agent = Agent(
    system_prompt=system_prompt,
    model=gemini_model,
    tools=[file_read, file_write, list_directory],  # Add your custom tools here
)


# Send a message to the agent
#response = local_agent("Who is Neil DeGrass Tyson?")
#print(response)

response = local_agent(r'''print the list of files from /content/sample_data/ in this local system''')
print(response)
```
